```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[23,45,67,89])
ax.set_title('Simple line plot using subplots')
ax.set_xlabel('Range')
ax.set_ylabel('Values',fontsize=18)
```

```python
x = np.random.random(20)
y = np.sin(x)
```

```python
fig, axs = plt.subplots(2, 2,figsize=(5,5))
axs[0,0].scatter(x,y)
axs[0,0].set_xlabel('A')
axs[0,0].set_ylabel('val1')
axs[0,0].set_title('Graph 1')

axs[0,1].plot(x,y,'--o')
axs[1,0].plot(x,y,'*c')
axs[1,1].bar(['A','B'],[100,230])
plt.suptitle('Total graphs together')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1, 2,sharex=False)
axs[0].plot(x,y)
axs[1].plot(x,y,'r')
plt.tight_layout()
```

```python
#plt.figure
##plt.figure(figsize=(8,8))
fig, axs = plt.subplots(1, 2,sharey=True,figsize=(6,2))
axs[0].plot(x,y)
axs[1].plot(x,y,'r')
plt.tight_layout()
```
