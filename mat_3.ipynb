# Pie plot

```python
# when we want %wise distribustion of a data
# univariate plot
# class vs %(out of 100%)
```

```python
# Covid 19 cases
states = ['MAH','KA','GUJ','AP']
count = [120,200,88,110]
```

```python
plt.pie(count,labels=states)
plt.show()
```

```python
# to get a % use autopct
plt.pie(count,labels=states,autopct='%.2f%%')
plt.show()
```

```python
# explod KA pie: means separate KA slice
plt.pie(count,labels=states,autopct='%.2f%%',explode=[0,0.2,0,0])
plt.show()
```
